<h2 style="color:blue;">  Using Surprise for Predicting Someone's Rating  </h2> 

### Create a dataset with a reversed rating
- Surprise assumes that 5 is the best and 1 is the lowest
- Our data assumes 1 is the best and 5 is lowest

```python
#Code Block 32

df_ap_brand_melt_1K.head()
```

```python
#Code Block 33

df_ap_brand_melt_1K_inv = df_ap_brand_melt_1K
df_ap_brand_melt_1K_inv['Rating'] = 6 - df_ap_brand_melt_1K_inv['Rating']
df_ap_brand_melt_1K_inv.head()
```

```python
#Code Block 34

reader = Reader(rating_scale=(1, 5)) #Sets the default scale to 1 to 5 - which is actually the default

#Specify User - Item - Rating by using the load_from_df function
data = Dataset.load_from_df(df_ap_brand_melt_1K_inv[['APID', 'Brand', 'Rating']], reader)
```

```python
#Code Block 35

train, test = train_test_split(data, test_size=0.25)
```

```python
%%time

#Code Block 36

from surprise import SVD, accuracy
algo = SVD()
algo.fit(train)
```

```python
%%time

#Code Block 37

predictions = algo.test(test)
```

```python
%%time

#Code Block 38

from surprise import accuracy
accuracy.rmse(predictions)
```

### What does a RMSE tell you?

- A RMSE of 0.8376 would mean the estimated ratings on average are about 0.8376 higher or lower than the actual ratings, on a 1 to 5 scale. Not great, not terrible.

```python
#Code Block 54

from collections import defaultdict
from surprise import SVD
from surprise import Dataset


def get_top_n(predictions, n=10):
    """Return the top-N recommendation for each user from a set of predictions.

    Args:
        predictions(list of Prediction objects): The list of predictions, as
            returned by the test method of an algorithm.
        n(int): The number of recommendation to output for each user. Default
            is 10.

    Returns:
    A dict where keys are user (raw) ids and values are lists of tuples:
        [(raw item id, rating estimation), ...] of size n.
    """

    # First map the predictions to each user.
    top_n = defaultdict(list)
    for uid, iid, true_r, est, _ in predictions:
        top_n[uid].append((iid, est))

    # Then sort the predictions for each user and retrieve the k highest ones.
    for uid, user_ratings in top_n.items():
        user_ratings.sort(key=lambda x: x[1], reverse=True)
        top_n[uid] = user_ratings[:n]

    return top_n


top_n = get_top_n(predictions, n=10)

# Print the recommended items for each user
for uid, user_ratings in top_n.items():
    print(uid, [iid for (iid, _) in user_ratings])
```
